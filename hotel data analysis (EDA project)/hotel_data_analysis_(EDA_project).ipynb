```python
sns.barplot( y = 'tip',x = 'day', data = data)
# this will show average tips by day in the form of barplot
```

```python
#by analyzing the plot shows the average tip amount for different days of the week
#The average tip amount is highest on Sundays and similar across the other days of the week
#This could be due to various factors, such as special occasions, relaxed weekend dining
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = data, hue = 'gender')
# this will show the total bill catagorised by gender in the form of violinplot
```

```python
#by analyzing the above plot-
#The violin shape represents the density distribution of the data.
#if the violin is wider ,then  more data points are present in that region.
#The black bar inside each violin represents the interquartile range (IQR), which contains 50% of the data.
#The white dot inside the bar represents the median
```
